# File handling and String manipulation

```python
# importing the necessary libraries

import numpy as np
```

```python
#reading the file in "text" variable

with open("conv.txt", "r") as f:
    text = f.read()

print(text)
```

<h3>Cleaning the sample conversation</h3>

<ul>
    <li>Remove unrecognized characters</li>
    <li>Removing the punctuation marks</li>
    <li>removing '\n' unicode</li>
    <li>seperating person name and their respective conversation</li>
</ul>

```python
encoded_text = text.encode("ascii", "ignore") #removing unrecognized characters
cleaned_text = encoded_text.decode("utf-8") #converting byte code text to normal text format

print(cleaned_text)
```

```python
# To remove the punctuation marks

punc = '''.?!,''' #creating a "punc" variable to store the puctuations

for element in cleaned_text:
    if element in punc:
        cleaned_text = cleaned_text.replace(element, " ") #removing the punctuations    
        
# text after removing punctuation
print(cleaned_text)
```

```python
# split the paragraphs into lines
para_v1 = cleaned_text.splitlines()

print(para_v1)
```

```python
para_v2 = []
characters = []

for lines in para_v1:
    sentence = lines.split(':') #splitting to seperate person and their respective conversations
    
    if len(sentence)!=1: #this ignores blank lines
        para_v2.append(sentence) # stores the sentences both person name and conversation
        characters.append(sentence[0]) # stores person names
```

Question 1. Find out the number of unique characters in the sample conversation?

Solution:

```python
# getting unique characters from the list
unique_characters = np.unique(characters)

# counting the number of unique characters
count_characters = len(unique_characters)

print(count_characters)
```

Question 2. Create a new text file for each of the characters with their name and store the unique words said by them in their respective file. Store one word in one line.

Solution:

```python
#extracting details for every unique characters present in the conversation
for people in unique_characters:
    speech_words=[]
    unique_words=[]

    # for each person their every conversation will be stored in speech_words
    for person, speech in para_v2:
        if people == person:
            speech_words+=(speech.lower()).split() # splitting after making it case insensitive
        unique_words=np.unique(speech_words) # extracting the unique words
  

    # creating a text file in append mode to store the unique words
    with open('{}.txt'.format(people), 'a') as f:
        for words in unique_words:
            f.write(words+'\n') # storing each word in new line

print("done!")
```
